```python
import os

def list_tree(path, max_depth=4, depth=0):
    if depth >= max_depth:
        return
    try:
        entries = sorted(os.listdir(path))
    except PermissionError:
        return
    for e in entries[:20]:
        full = os.path.join(path, e)
        prefix = '  ' * depth
        if os.path.isdir(full):
            print(f'{prefix}{e}/')
            list_tree(full, max_depth, depth + 1)
        else:
            print(f'{prefix}{e}')
    if len(entries) > 20:
        print(f'{"  " * depth}... and {len(entries) - 20} more')

list_tree('/kaggle/input')
```
